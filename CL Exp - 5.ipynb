```python
import pandas as pd
import numpy as np

df = pd.read_csv('CovidIndiaData-01-04-2020.csv')
pd.read_csv('CovidIndiaData-01-04-2020.csv')
```

```python
print("The mean of Confirmed:", df["Confirmed"].mean())
print("The mean of Deaths:", df["Deaths"].mean())
print("The mean of Recovered:", df["Recovered"].mean())
print("The mean of week:", df["week"].mean())
```

```python
print("The median of Confirmed:", df["Confirmed"].median())
print("The median of Deaths:", df["Deaths"].median())
print("The median of Recovered:", df["Recovered"].median())
print("The median of week:", df["week"].median())
```

```python
print("The mode of Confirmed:", df["Confirmed"].mode())
print("The mode of Deaths:", df["Deaths"].mode())
print("The mode of Recovered:", df["Recovered"].mode())
print("The mode of week:", df["week"].mode())
```

```python
print("The standard deviation of Confirmed:", df["Confirmed"].std())
print("The standard deviation of Deaths:", df["Deaths"].std())
print("The standard deviation of Recovered:", df["Recovered"].std())
print("The standard deviation of week:", df["week"].std())
```

```python
print("The variance of Confirmed:", df["Confirmed"].var())
print("The variance of Deaths:", df["Deaths"].var())
print("The variance of Recovered:", df["Recovered"].var())
print("The variance of week:", df["week"].var())
```
