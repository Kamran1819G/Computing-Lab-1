```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a dataset
tips = sns.load_dataset("tips")

# Create a scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
sns.lineplot(x="timepoint", y="signal", hue="region", data=fmri)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a dataset
titanic = sns.load_dataset("titanic")

# Create a bar plot
sns.barplot(x="age", y="fare", data=titanic)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a dataset
iris = sns.load_dataset("iris")

# Create a histogram
sns.histplot(data=iris, x="sepal_length")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a dataset
flights = sns.load_dataset("flights").pivot("month", "year", "passengers")

# Create a heatmap
sns.heatmap(flights)

# Show the plot
plt.show()
```
